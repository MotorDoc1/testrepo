```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("PythonWeek5.csv")
fig,ax=plt.subplots()
my_scatter_plot = ax.scatter(df[" Net Score"], df["Number of Retweets"])
plt.show
ax.set_title('Number of Retweets vs the Net Score')
ax.set_xlabel('Net Score')
ax.set_ylabel('Retweets')
```
